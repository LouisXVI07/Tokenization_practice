```python
def Byt(token):
    count={}
    for pair in zip(token, token[1:]):
        count[pair]=count.get(pair,0)+1
    return count
```

```python
stats=(Byt(tokens))
```

```python
addre={}
idx=256
while (1):
    stats=(Byt(tokens))
    top_pair=max(stats,key=stats.get)
    if (stats[top_pair]<=5):
        break
    addre[top_pair]=idx
    
    i=0
    temp=[]
    while (i<len(tokens)):
        if i < len(tokens)-1 and tokens[i]==top_pair[0] and tokens[i+1]==top_pair[1]:
            #print(f"merging {top_pair} into {idx}")
            temp.append(idx)
            i=i+2
            
        else:
            temp.append(tokens[i])
            i=i+1
    idx=idx+1
    tokens=temp
print(len(tokens))
```

```python
def decoder(tokens, addre):
   
    merges = list(addre.items())[::-1]   

    for pair, mid in merges:
        k = 0
        while k < len(tokens):
            if tokens[k] == mid:
                
                tokens[k] = pair[0]
                tokens.insert(k + 1, pair[1])
                k += 2   
            else:
                k += 1

    return tokens
```

```python
def encode(txt,addre):
    tokens=list(txt.encode("utf-8"))
    
    for i,j in addre.items():
        k1=i[0]
        k2=i[1]
        m=len(tokens)
        t=0
        temp=[]
        while (t<m):
            if t < len(tokens)-1 and tokens[t]==k1 and tokens[t+1]==k2:
            #print(f"merging {top_pair} into {idx}")
                temp.append(j)
                t=t+2
            
            else:
                temp.append(tokens[t])
                t=t+1
        tokens=temp
        
    return tokens          
```

```python
with open("toy.txt",'w',encoding="utf-8") as f:
    f.write ("""SentencePiece is an open-source, language-independent subword tokenizer and detokenizer designed for neural network-based text processing systems, such as neural machine translation (NMT) and natural language processing (NLP) tasks.
 It enables end-to-end text processing by training subword models directly from raw sentences without requiring language-specific pre-tokenization, making it highly versatile for diverse languages, including those with limited resources.
 SentencePiece supports multiple subword algorithms, including byte-pair-encoding (BPE) and unigram language models, and allows users to specify a predetermined vocabulary size to balance model efficiency and expressiveness.
 It treats input text as a sequence of Unicode characters, preserving whitespace through a special symbol (e.g., "▁") to ensure reversible tokenization and detokenization, and includes customizable normalization (e.g., NFKC) and direct id generation for seamless integration with neural models""")
```

```python
import os

options = dict(
  # input spec
  input="toy.txt",
  input_format="text",
  # output spec
  model_prefix="tok400", # output filename prefix
  # algorithm spec
  # BPE alg
  model_type="bpe",
  vocab_size=400,
  # normalization
  normalization_rule_name="identity", # ew, turn off normalization
  remove_extra_whitespaces=False,
  input_sentence_size=200000000, # max number of training sentences
  max_sentence_length=4192, # max number of bytes per sentence
  seed_sentencepiece_size=1000000,
  shuffle_input_sentence=True,
  # rare word treatment
  character_coverage=0.99995,
  byte_fallback=True,
  # merge rules
  split_digits=True,
  split_by_unicode_script=True,
  split_by_whitespace=True,
  split_by_number=True,
  max_sentencepiece_length=16,
  add_dummy_prefix=True,
  allow_whitespace_only_pieces=True,
  # special tokens
  unk_id=0, # the UNK token MUST exist
  bos_id=1, # the others are optional, set to -1 to turn off
  eos_id=2,
  pad_id=-1,
  # systems
  num_threads=os.cpu_count(), # use ~all system resources
)

spm.SentencePieceTrainer.train(**options)
```

```python
sp = spm.SentencePieceProcessor()
sp.load('tok400.model')
vocab = [[sp.id_to_piece(idx), idx] for idx in range(sp.get_piece_size())]
vocab
```
